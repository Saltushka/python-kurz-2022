```python
import pandas as pd
import matplotlib.pyplot as plt
file_path = 'platy_2021_02.csv'
```

```python
data = pd.read_csv(file_path)
```

```python
num_bins = 20
```

Vytvoř histogram

```python
plt.hist(data['plat'], bins=num_bins, color='blue', edgecolor='black')
```

```python
plt.xlabel('Platy')
plt.ylabel('Počet zaměstnanců')
plt.title('Rozložení platů v softwarové firmě (únor 2021)')
```

Zobrazení grafu

```python
plt.show()
```
