```python
import pandas as pd
import seaborn as sns
data = pd.read_csv("temperature.csv")
data.head()
```

```python
data = data[data["City"].isin(["Helsinki", "Miami Beach", "Tokyo"])]
data.plot(kind="box", column="AvgTemperature", by="City")
```

```python
data = data[data["City"].isin(["Helsinki", "Miami Beach", "Tokyo"])]
sns.boxplot(data, y="AvgTemperature", x="City")
```

```python
data_pivot = pd.pivot(data, index="Day", columns="City", values="AvgTemperature")
data_pivot
```

```python
data_pivot.plot(kind="box")
```
